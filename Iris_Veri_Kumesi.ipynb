# Regresyon Algorithms  

### Label Encoder

Regresyon yöntemlerinde numerik değerler işler, kategorik değerler için sınıflandırma algoritmaları kullanılır. Bu sebeple aşağıda regresyon algoritmalarında iris kolununu `LabelEncoder` ile numeric değerlere dönüştürürüz.

Ayrıca, Yukarıda çözümlenen algoritmalar Sınıflandırma algoritması, aşağıda çözümlenen algoritmalar regresyon algoritmasıdır.
Sınıflandırma algoritmalarında doğruluk ölçümü için `confusion_matrix` kullanılırken,
Regresyon algoritmalarında doğruluk ölçümü için `r^2_score ` kullanılır.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
new_iris
```

## 3D Çizim x ölçekli

```python
#import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(7, 7))
ax = fig.add_subplot( projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(df)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```
