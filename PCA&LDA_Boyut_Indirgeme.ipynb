```python
x= dataFrame.drop("Customer_Segment", axis=1).values
y= dataFrame["Customer_Segment"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.33, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
x_train= scaler.fit_transform(x_train)
```

```python
x_test= scaler.fit_transform(x_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
X_train= pca.fit_transform(x_train)
```

```python
X_test= pca.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression(random_state=0)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred= lr.predict(x_test)
```

```python
lrpca= LogisticRegression(random_state=0)
```

```python
lrpca.fit(X_train, y_train)
```

```python
y_predpca= lrpca.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm= confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
cmpca= confusion_matrix(y_test, y_predpca)
```

```python
cmpca
```

```python
cm_= confusion_matrix(y_pred, y_predpca)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda= LDA(n_components=2)
```

```python
X_train_lda = lda.fit_transform(x_train, y_train)
```

```python
X_test_lda= lda.transform(x_test)
```

```python
lr_lda=LogisticRegression(random_state=0)
```

```python
lr_lda.fit(X_train_lda, y_train)
```

```python
y_pred_lda= lr_lda.predict(X_test_lda)
```

```python
cm_lda= confusion_matrix(y_test, y_pred_lda)
```
