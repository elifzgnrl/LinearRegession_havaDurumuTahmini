```python
x=dataFrame.drop("Type", axis=1).values
y=dataFrame["Type"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train =scaler.transform(x_train)
```

```python
x_test= scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add (Dense(units= 30, activation="relu"))
model.add (Dense(units= 15, activation="relu"))
model.add (Dense(units= 15, activation="relu"))
model.add (Dense(units= 1, activation="sigmoid")) #0 ve 1 aralığında değer veriyor

model.compile(loss="binary_crossentropy", optimizer= "adam") #activation function sigmoid olduğu için loss kayıp fonksiyonu bool değer alan binary_crossentropy fonksiyonunun uygun bulduk.
```

```python
model.fit(x= x_train, y= y_train, epochs=700, validation_data=(x_test, y_test), verbose=1)
```

```python
model = Sequential()
model.add (Dense(units= 30, activation="relu"))
model.add (Dense(units= 15, activation="relu"))
model.add (Dense(units= 15, activation="relu"))
model.add (Dense(units= 1, activation="sigmoid")) #0 ve 1 aralığında değer veriyor

model.compile(loss="binary_crossentropy", optimizer= "adam")
```

```python
earlyStoppping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25 )
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data= (x_test, y_test), verbose=1, callbacks=[earlyStoppping])
```

```python
modelKaybi= pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model= Sequential()
model.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data= (x_test, y_test), verbose=1, callbacks=[earlyStoppping])
```

```python
kayipDF= pd.DataFrame(model.history.history)
```

```python
kayipDF.plot()
```

```python
predicts= model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicts))
```
