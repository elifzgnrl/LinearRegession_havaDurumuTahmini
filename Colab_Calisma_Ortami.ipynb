`Numpy` ve `Matplotlib` kütüphaneleri kullanılacaktır.

```python
import numpy as np
from matplotlib import pyplot as plt

ys= 50 + np.random.randn(50)
x = [ x for x in range (len (ys)) ]
plt.plot(x,ys,'*')
plt.fill_between( x , ys , 45 , where = ( ys > 45 ), facecolor='m', alpha= 0.5)
#plt.fill_between( x , ys , 45, facecolor='m', alpha= 0.5)
#plt.fill_between( x , ys , 50 , where = ( ys > 50 ), facecolor='m')
#plt.fill_between( x , ys , 10 , where = ( ys > 10 ), facecolor='m', alpha= 0.5)
#plt.fill_between( x , ys , 50 , where = ( ys > 50 ), facecolor='m', alpha= 0.5)

plt.title('Örnek Gösterim')
plt.show()
```
